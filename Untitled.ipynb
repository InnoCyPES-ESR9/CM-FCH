```python
# Import libraries
from PIL import Image
import cv2
import numpy as np
import requests
```

```python
image = Image.open(r'C:\Users\ksu\Desktop\Congestion monitoring fast charging hubs\data\denmark_road_cameras\DK_20220315110000_716.jpg')
image = image.resize((450,250))
image_arr = np.array(image)
image
```

```python
grey = cv2.cvtColor(image_arr,cv2.COLOR_BGR2GRAY)
Image.fromarray(grey)
```

```python
blur = cv2.GaussianBlur(grey,(5,5),0)
Image.fromarray(blur)
```

```python
dilated = cv2.dilate(blur,np.ones((3,3)))
Image.fromarray(dilated)
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (2, 2))
closing = cv2.morphologyEx(dilated, cv2.MORPH_CLOSE, kernel) 
Image.fromarray(closing)
```

```python
#car_cascade = cv2.CascadeClassifier('cars.xml')

car_cascade = cv2.CascadeClassifier(r'C:\Users\ksu\Anaconda3\envs\cmfch\Lib\site-packages\cv2\data\cars.xml')
cars = car_cascade.detectMultiScale(grey, 1.1, 1)
#cars



#eye_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_eye.xml')
```

```python
cars
```

```python
cnt = 0
for (x,y,w,h) in cars:
    cv2.rectangle(image_arr,(x,y),(x+w,y+h),(255,0,0),2)
    cnt += 1
print(cnt, " cars found")
Image.fromarray(image_arr)
```
